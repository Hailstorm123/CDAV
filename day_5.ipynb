```python
test_case = ["ann" , "civic" , "level" , "hannah"]
for item in test_case :
  cleaned =item .strip("'/'"). lower ()
  is_palindrome = True
  left =0
  right = len(cleaned) -1
  while left < right:
    if cleaned [left] !=cleaned [right] :
      is_palindrome= False
      break
    left += 1
    right -= 1
  if is_palindrome:
    print(f"'{item}' -> Is a palindrome !")
  else :
    print(f"'{item}' -> Is NOT a palindrome .")
```

```python
m=int(input("enter first positive integer "))
n= int(input("enter second positive integer "))
d= min (m,n)
while m % d !=0 or n % d !=0:
  d -= 1
print (f"greatest common divisor of {m} and {n} is {d} .")
```

```python
message = input("enter message to encode :")
shift =int (input("enter shift value (positive or ne):"))
encoded_message =""
for char in message :
  if "a" < char  and char <  "z" :
    pos = ord (char) - ord ("a")
    new_pos = (pos + shift) % 26
    new_char = chr (new_pos + ord ("a"))
    encoded_message += new_char
if "A" <= char <= "Z" :
   pos = ord (char) - ord ("A")
   new_pos = (pos + shift) % 26
   new_char = chr (new_pos + ord ("A"))
   encoded_message += new_char
print (f"encoded message is : {encoded_message} .")
```
